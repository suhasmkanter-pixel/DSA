```python
#Check if Array is Sorted
a = [11,10,9,5]



def check_array(a): 
    ascending = True 
    descending = True
    for i in range(1,len(a)):  
        if a[i-1] > a[i]: 
            ascending =  False
        if a[i-1] < a[i]:
            descending = False
    return ascending or descending  
    
if check_array(a): 
    print("THe array is Sorted")
else:
    print("Not sorted")
```

```python
# Left Rotate Array by One Place
def rotate_array(arr,rotation):
    
    return arr[rotation:] + arr[:rotation]
        


         
   
 
arr = [1, 2, 3, 4, 5]
rotate = 2
rotate_array(arr,rotate)
```
